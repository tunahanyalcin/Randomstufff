```
import pandas as pd
import numpy as np
import time
from google.colab import files
uploaded = files.upload() #This row is for uploading the excel file to colab, if you use smtg else, you might not need it
A = np.zeros((73,5))

''' matrix A:
1st column: Counts the number of nan values
2nd column: Counts the number of positive values
3rd column: Aggregation of the positive values
4th column: Counts the number of negative values
5th column: Aggregation of the negative values

Each row represents a participant
'''

dataset = pd.read_excel('babip_test_2.xlsx')        #Reading the file
dataset_2 = dataset.reset_index()                   #Indexing the rows
#print(dataset_2.head())                            #This prints out the first 5 rows of the data

for i in range(73):
  dataset_3 = dataset_2.loc[i, :]                   #This selects one row(i) and all the columns for each loop  
  for j in dataset_3:                               #Columns are defined as 'j'
    if j > 0:
      A[i][1] += 1                                  #A[x][y] selects xth row and yth column of a matrix
      A[i][2] += j
    elif j < 0:
      A[i][3] += 1
      A[i][4] += j
    else:
      A[i][0] += 1

ans =input('What is 21x20?')
if ans == 420:
  print('You smart thang')
else:
  print('Time for more kindergarten :(')
time.sleep(2)
print('Welcome! You will see your result in 10secs...')
time.sleep(3)
print('Waiting for it just makes it better...')
time.sleep(3)
print('I know I know. Just a little more...')
time.sleep(1)
print('3')
time.sleep(1)
print('2')
time.sleep(1)
print('1')
time.sleep(1)

print(A)

print('Thanks for choosing us!')
```
